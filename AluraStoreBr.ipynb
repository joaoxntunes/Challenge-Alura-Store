### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
df = pd.concat([loja, loja2, loja3, loja4])
faturamento_por_loja = df.groupby("Local da compra")["Preço"].sum().reset_index()
```

```python
faturamento_por_loja
```

# 2. Vendas por Categoria


```python
vendas_por_categoria = pd.crosstab(df["Local da compra"], df["Categoria do Produto"])
```

```python
vendas_por_categoria
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao = df.groupby("Local da compra")["Avaliação da compra"].mean().reset_index()
```

```python
media_avaliacao
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_por_produto = df.groupby(['Local da compra', 'Produto']).size().reset_index(name='Quantidade_vendida')
mais_vendidos = vendas_por_produto.loc[vendas_por_produto.groupby('Local da compra')['Quantidade_vendida'].idxmax()]
menos_vendidos = vendas_por_produto.loc[vendas_por_produto.groupby('Local da compra')['Quantidade_vendida'].idxmin()]
```

```python
print("Produtos mais vendidos por loja:")
print(mais_vendidos)

print("\nProdutos menos vendidos por loja:")
print(menos_vendidos)
```

# 5. Frete Médio por Loja


```python
frete_medio = df.groupby("Local da compra")["Frete"].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

# 1. Faturamento por loja
faturamento_por_loja = df.groupby("Local da compra")["Preço"].sum().reset_index()

plt.figure(figsize=(8,5))
plt.bar(faturamento_por_loja["Local da compra"], faturamento_por_loja["Preço"], color='skyblue')
plt.title("Faturamento por Loja")
plt.xlabel("Loja")
plt.ylabel("Faturamento")
plt.show()


# 2. Média de avaliação por loja
media_avaliacao = df.groupby("Local da compra")["Avaliação da compra"].mean().reset_index()

plt.figure(figsize=(8,5))
plt.plot(media_avaliacao["Local da compra"], media_avaliacao["Avaliação da compra"], marker='o', color='green')
plt.title("Média de Avaliação por Loja")
plt.xlabel("Loja")
plt.ylabel("Média de Avaliação")
plt.ylim(0,5)  # assumindo que a nota máxima é 5
plt.grid(True)
plt.show()


# 3. Produtos mais vendidos por loja
vendas_por_produto = df.groupby(['Local da compra', 'Produto']).size().reset_index(name='Quantidade_vendida')
mais_vendidos = vendas_por_produto.loc[vendas_por_produto.groupby('Local da compra')['Quantidade_vendida'].idxmax()]

plt.figure(figsize=(10,6))
for i, loja_nome in enumerate(mais_vendidos['Local da compra'].unique()):
    subset = mais_vendidos[mais_vendidos['Local da compra'] == loja_nome]
    plt.barh(subset['Produto'], subset['Quantidade_vendida'], label=loja_nome)

plt.title("Produtos Mais Vendidos por Loja")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Produto")
plt.legend()
plt.show()
```

## Relatório

O objetivo deste relatório é analisar o desempenho das lojas em diferentes estados para identificar a melhor opção de investimento para o Senhor João.

**Loja de SP possui o maior faturamento superando até 2,5 milhões.**

**Também possui alta média de avaliação dos clientes, próxima de 4.5.**

**Os produtos mais vendidos (como micro-ondas, lavadora e assistente virtual), possui forte presença na loja de SP.**

**Embora as lojas do Rio de Janeiro e Minas Gerais apresentem bom desempenho, seus faturamentos e avaliações não superam os resultados observados em São Paulo.**

**Os produtos menos vendidos (como corda de pular, faqueiro) tem baixa relevância em todas as Lojas.**

Diante desses indicadores, a loja de São Paulo representa o melhor retorno financeiro e a opção mais atrativa para potenciais compradores
